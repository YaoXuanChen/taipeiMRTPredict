```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
datas=[]
# 爬取109年至111年旅運量資料
for i in range(109,112):
#     爬取1月至12月資料
    for j in range(1,13):
        if j < 10:
            url=f'https://web.metro.taipei/RidershipCounts/c/{i}0{j}.htm'
        else:
            url=f'https://web.metro.taipei/RidershipCounts/c/{i}{j}.htm'
        s=Service(r'/Users/chenyaoxuan/Desktop/tools/chromedriver/chromedriver')
        chrome=webdriver.Chrome(service=s)
        chrome.get(url)
        time.sleep(5)

        soup=BeautifulSoup(chrome.page_source,'lxml')
        
        date=1
#         擷取每日旅運量資料
        try:
            for k in soup.find('tbody').find_all('tr')[2:-4]:
                passengers=int(k.find('td',align="right").text.strip().replace(',',''))
                if date < 10:
                    datas.append([f'{i}年{j}月0{date}日',passengers])
                    print(f'{i}年{j}月0{date}日{passengers}')
                    date+=1
                else:
                    datas.append([f'{i}年{j}月{date}日',passengers])
                    print(f'{i}年{j}月{date}日{passengers}')
                    date+=1
        except:
            print('None')
        finally:
            chrome.quit()
```

```python
# 將資料匯出為csv檔
pd.DataFrame(datas,columns=['date','passengers']).to_csv('totalPassenger.csv',index=False,encoding='utf-8-sig')
```
